```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = np.array(Image.open('bridge.jpg'))
plt.imshow(img, interpolation='nearest')
```

```python
gray = np.zeros((img.shape[0], img.shape[1]))

for i in range(len(img)):
    for j in range(len(img[i])):
        gray[i][j] = 0.299*img[i][j][0] + 0.587*img[i][j][1] + 0.114*img[i][j][2]

plt.imshow(gray, interpolation='nearest', cmap=plt.get_cmap('gray'))
plt.show()
```

```python
#Median Blur

from statistics import median

kernel_size = 3
medianBlur = np.zeros((gray.shape[0]-kernel_size+1, gray.shape[1]-kernel_size+1))

for i in range(gray.shape[0]-kernel_size+1):
    for j in range(gray.shape[1]-kernel_size+1):
        
        medianBlur[i][j] = median(gray[i:i+kernel_size, j:j+kernel_size].reshape(1, -1)[0])

plt.imshow(medianBlur, interpolation='nearest', cmap=plt.get_cmap('gray'))
plt.show()
```
