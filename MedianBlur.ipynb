```python
import numpy as np
from PIL import Image as image
from IPython.display import Image, display
from io import BytesIO

img = np.array(image.open('gal.png'))
f = BytesIO()
image.fromarray(img.astype('uint8')).save(f, 'png')
display(Image(f.getvalue()))
```

```python
gray = np.zeros((img.shape[0], img.shape[1]))

for i in range(len(img)):
    for j in range(len(img[i])):
        gray[i][j] = 0.299*img[i][j][0] + 0.587*img[i][j][1] + 0.114*img[i][j][2]

f = BytesIO()
image.fromarray(gray.astype('uint8')).save(f, 'png')
display(Image(f.getvalue()))
```

```python
#Median Blur

from statistics import median

kernel_size = 3
medianBlur = np.zeros((gray.shape[0]-kernel_size+1, gray.shape[1]-kernel_size+1))

for i in range(gray.shape[0]-kernel_size+1):
    for j in range(gray.shape[1]-kernel_size+1):
        
        medianBlur[i][j] = median(gray[i:i+kernel_size, j:j+kernel_size].reshape(1, -1)[0])

f = BytesIO()
image.fromarray(medianBlur.astype('uint8')).save(f, 'png')
display(Image(f.getvalue()))
```
