```python
import numpy as np
from PIL import Image as image
from IPython.display import display, Image
from io import BytesIO

img = np.array(image.open('monument.jpg'))
f = BytesIO()
image.fromarray(img.astype('uint8')).save(f, 'jpeg')
display(Image(f.getvalue()))
```

```python
gray = np.zeros((img.shape[0], img.shape[1]))

for i in range(len(img)):
    for j in range(len(img[i])):
        gray[i][j] = 0.299*img[i][j][0] + 0.587*img[i][j][1] + 0.114*img[i][j][2]

f = BytesIO()
image.fromarray(gray.astype('uint8')).save(f, 'jpeg')
display(Image(f.getvalue()))
```

```python
#### Box Blur ####

radius = 3
box_blur = np.zeros((gray.shape[0]-2, gray.shape[1]-2))
kernel = np.ones((radius, radius))

for i in range(gray.shape[0]-8):
    for j in range(gray.shape[1]-8):
        
        total = np.sum(np.multiply(gray[i:i+radius, j:j+radius], kernel))
        box_blur[i][j] = total/(radius**2)

f = BytesIO()
image.fromarray(box_blur.astype('uint8')).save(f, 'jpeg')
display(Image(f.getvalue()))
```
