```python
import cv2
from PIL import Image
import numpy as np

img = cv2.imread('low.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Image.fromarray(img)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(img.flatten(), range(0, 256))
plt.title('Histogram plot')
plt.show()
```

```python
h, w= img.shape
num_bins = 256
num_pixels = h*w
```

```python
img
```

```python
from collections import Counter

count = Counter(img.flatten())
count
```

```python
P = [0]*num_bins

for i in range(num_bins):
    if count.get(i):
        P[i] = count[i]
P
```

```python
cuf = [0]*num_bins
cuf[0] = P[0]

for i in range(1, num_bins):
    cuf[i] = cuf[i-1] + P[i]

cuf
```

```python
plt.plot(range(0, 256), cuf)
plt.title('Cumulative Distribution plot')
plt.show()
```

```python
eh = [0]*num_bins
cufMin = min(cuf)

for i in range(256):
    eh[i] = round((cuf[i] - cufMin)/(num_pixels - cufMin) * (num_bins-1))
```

```python
equalizedImage = np.zeros((h, w))

for i in range(h):
    for j in range(w):
        equalizedImage[i][j] = eh[img[i][j]]
```

```python
plt.hist(equalizedImage.flatten(), range(0, 256))
plt.title('Equalized Histogram plot')
plt.show()
```

```python
count1 = Counter(equalizedImage.flatten())
count1
```

```python
P1 = [0]*num_bins

for i in range(num_bins):
    if count1.get(i):
        P1[i] = count1[i]
P1
```

```python
cuf1 = [0]*num_bins
cuf1[0] = P1[0]

for i in range(1, num_bins):
    cuf1[i] = cuf1[i-1] + P1[i]

cuf1
```

```python
plt.plot(range(0, 256), cuf1)
plt.title('Equalized Cumulative Distribution plot')
plt.show()
```

```python
Image.fromarray(equalizedImage).convert('RGB')
```
